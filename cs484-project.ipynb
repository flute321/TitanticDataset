```python
# Load train dataset on local device
train = pd.read_csv('train.csv')
test = pd.read_csv("test.csv")
train.head()
```

```python
#trying to figure the amount of nan values
count = train.isna().sum()
count 
```

```python
#trying to figure the amount of nan values
count = test.isna().sum()
count
```

```python
#replacing all nan values of age with mean age for train
mean_age = train["Age"].mean()
train['Age'].fillna(mean_age, inplace=True)
count = train.isna().sum()
count
```

```python
#replacing all nan values of age with mean age for test
mean_age = test["Age"].mean()
test['Age'].fillna(mean_age, inplace=True)
count = test.isna().sum()
```

```python
#replacing all nan values embark with the mode in train
mode_embark = train['Embarked'].mode()[0]  
train['Embarked'].fillna(mode_embark, inplace=True)
count = train.isna().sum()
count
```

```python
#replacing all nan values of age with mean age for test
mean_age = test["Fare"].mean()
test['Fare'].fillna(mean_age, inplace=True)
count = test.isna().sum()
count
```

```python
unique_ticket = len(train['Ticket'].unique())
unique_ticket
```

```python
#creating a new column for age
def modify_age(row):
    original_value = row['Age']
    if original_value < 11:
        return "Under 12 years old"  
    elif original_value >= 12 and original_value < 24:
        return "12-24 years old"   
    elif original_value >= 24 and original_value < 45:
        return "24-45 years old"  
    elif original_value >= 45 and original_value < 65:
        return "45-64 years old"   
    else:
        return "65 years or older" 

train['Ages'] = train.apply(modify_age, axis=1)
test['Ages'] = test.apply(modify_age, axis=1)

train.head()
```

```python
#creating a new column for sibsp
def modify_sibsp(row):
    original_value = row['SibSp']
    if original_value == 0:
        return "0 sibsp members"  
    elif original_value >= 1 and original_value < 5:
        return "1-4 sibsp members"    
    else:
        return "5+ sibsp members" 

train['Sibsp'] = train.apply(modify_sibsp, axis=1)
test['Sibsp'] = test.apply(modify_sibsp, axis=1)

train.head()
```

```python
#creating a new column for sibsp
def modify_parch(row):
    original_value = row['Parch']
    if original_value == 0:
        return "0 parch members"  
    elif original_value >= 1 and original_value < 5:
        return "1-4 parch members"    
    else:
        return "5+ parch members" 

train['parch'] = train.apply(modify_parch, axis=1)
test['parch'] = test.apply(modify_parch, axis=1)

train.head()
```

```python
#creating a new column for sibsp
def modify_fare(row):
    original_value = row['Fare']
    if original_value < 10:
        return "Fare Under 10"  
    elif original_value >= 11 and original_value < 30:
        return "Fare 11-30"    
    else:
        return "Fare 31+" 

train['fare'] = train.apply(modify_fare, axis=1)
test['fare'] = test.apply(modify_fare, axis=1)

train.head()
```

```python
#dropping name and ticket(cuz unique and doesnt help with data analysis)
#dropping cabin cuz 75% values are nan
train = train.drop(['Name', 'Cabin', 'Ticket', 'PassengerId', 'SibSp', 'Parch', 'Fare', 'Age'], axis=1)
test = test.drop(['Name', 'Cabin', 'Ticket', 'PassengerId', 'SibSp', 'Parch', 'Fare', 'Age'], axis=1)
train.head()
```

```python
# Get features and labels
features = train.values[:, 1:]
labels = train.values[:, 0]
print('array of labels: shape ' + str(np.shape(labels)))
print('array of feature matrix: shape ' + str(np.shape(features)))
labels[0]
```

```python
from sklearn.preprocessing import OneHotEncoder

#encode the features
encoder= OneHotEncoder(sparse =False)
features_transformed = encoder.fit_transform(features)

print('array of x_train: shape ' + str(np.shape(features_transformed)))
print(features_transformed[0])
```

```python
#making the training and validation set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_transformed, labels, test_size=0.2, random_state=42)
print('array of x_train: shape ' + str(np.shape(X_train)))
print('array of x_test: shape ' + str(np.shape(X_test)))
print('array of y_test: shape ' + str(np.shape(y_train)))
print('array of y_train: shape ' + str(np.shape(y_test)))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.linear_model import LogisticRegression
model_2 = LogisticRegression()
model_2.fit(X_train, y_train)
y_pred_2 = model_2.predict(X_test)

score = model_2.score(X_test, y_test)
print(score)
```
