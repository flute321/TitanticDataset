```python
#replacing all nan values of age with mean age for train
mean_age = train["Age"].mean()
train['Age'].fillna(mean_age, inplace=True)
count = train.isna().sum()
count
```

```python
#replacing all nan values of age with mean age for test
mean_age = test["Age"].mean()
test['Age'].fillna(mean_age, inplace=True)
count = test.isna().sum()
```

```python
#replacing all nan values of age with mean age for test
mean_age = test["Fare"].mean()
test['Fare'].fillna(mean_age, inplace=True)
count = test.isna().sum()
count
```

```python
#creating a new column for age
def modify_age(row):
    original_value = row['Age']
    if original_value < 11:
        return "Under 12 years old"  
    elif original_value >= 12 and original_value < 24:
        return "12-24 years old"   
    elif original_value >= 24 and original_value < 45:
        return "24-45 years old"  
    elif original_value >= 45 and original_value < 65:
        return "45-64 years old"   
    else:
        return "65 years or older" 

train['Ages'] = train.apply(modify_age, axis=1)
test['Ages'] = test.apply(modify_age, axis=1)

train.head()
```

```python
#creating a new column for sibsp
def modify_sibsp(row):
    original_value = row['SibSp']
    if original_value == 0:
        return "0 sibsp members"  
    elif original_value >= 1 and original_value < 5:
        return "1-4 sibsp members"    
    else:
        return "5+ sibsp members" 

train['Sibsp'] = train.apply(modify_sibsp, axis=1)
test['Sibsp'] = test.apply(modify_sibsp, axis=1)

train.head()
```

```python
#creating a new column for sibsp
def modify_parch(row):
    original_value = row['Parch']
    if original_value == 0:
        return "0 parch members"  
    elif original_value >= 1 and original_value < 5:
        return "1-4 parch members"    
    else:
        return "5+ parch members" 

train['parch'] = train.apply(modify_parch, axis=1)
test['parch'] = test.apply(modify_parch, axis=1)

train.head()
```

```python
#creating a new column for sibsp
def modify_fare(row):
    original_value = row['Fare']
    if original_value < 10:
        return "Fare Under 10"  
    elif original_value >= 11 and original_value < 30:
        return "Fare 11-30"    
    else:
        return "Fare 31+" 

train['fare'] = train.apply(modify_fare, axis=1)
test['fare'] = test.apply(modify_fare, axis=1)

train.head()
```

```python
#dropping name and ticket(cuz unique and doesnt help with data analysis)
#dropping cabin cuz 75% values are nan
train = train.drop(['Name', 'Cabin', 'Ticket', 'PassengerId', 'SibSp', 'Parch', 'Fare', 'Age'], axis=1)
test = test.drop(['Name', 'Cabin', 'Ticket', 'PassengerId', 'SibSp', 'Parch', 'Fare', 'Age'], axis=1)
train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

#encode the features
encoder= OneHotEncoder(sparse =False)
features_transformed = encoder.fit_transform(features)

print('array of x_train: shape ' + str(np.shape(features_transformed)))
print('array of y_train: shape ' + str(np.shape(labels)))
print(labels[0])

print(type(features_transformed))
```
